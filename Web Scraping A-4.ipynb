## 1. Scrape the details of most viewed videos on YouTube from Wikipedia: Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos. You need to find following details:Rank, Name, Artist, Upload Date, Views 

```python
# Function Definition
def youtube_videos(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(10)
    
#     Let's create empty lists to store necessary information
    rank = []
    name = []
    artist = []
    upload_date = []
    views = []
    
#     extracting rank
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]"):
        rank.append(i.text)
#     extracting name
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]"):
        name.append(i.text)
#     extracting artist
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]"):
        artist.append(i.text)
#     extracting upload_date
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]"):
        upload_date.append(i.text)
#     extracting views
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]"):
        views.append(i.text)
    
    df = pd.DataFrame({})
    df['Rank']=rank[:30]
    df['Name']=name[:30]
    df['Artist']=artist[:30]
    df['Upload Date']=upload_date[:30]
    df['Views(Billions)']=views[:30]
    
    driver.close()
    
    return df


# Calling Function
youtube_videos('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

## 2. Scrape the details of team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details: Match Title, Series, Place, Date, Time. Note: - From bcci.tv home page you have to reach to the international fixtures page through code.  

```python
# Function Definition
def cricket_fixtures(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
    international = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")
    international.click()
    time.sleep(4)
    fixtures = driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']")
    fixtures.click()
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'js-list')))
    
#     extracting all the cricket matches url
    match_href = []
    for i in driver.find_elements_by_xpath("//div[@class='js-list']/a"):
        match_href.append(i.get_attribute('href'))
    
#     Let's create empty lists to store necessary information
    match_title = []
    series = []
    place=[]
    date = []
    mat_time = []
    
    for url in match_href:
        driver.get(url)
        delay=10
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'mc-header-info__match-description')))
        
#         extracting match_title
        mt = driver.find_element_by_xpath("//p[@class='mc-header-info__match-description']")
        match_title.append(mt.text.replace(' VIEW SERIES',''))
        
#         extracting series
        s = driver.find_element_by_xpath("//h1[@class='mc-header-info__title']/strong")
        series.append(s.text)
        
#         extracting place
        md = [] # Match Details
        for i in driver.find_elements_by_xpath("//div[@class='match-info__details']/ul/li[3]/span[2]"):
            md.append(i)
        place.append(md[1].text)
        
#         extracting date
        d = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/strong")
        date.append(d.text)
        
#         extracting match time
        t = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/span[1]")
        mat_time.append(t.text)

    df = pd.DataFrame({})
    df['Match_Title']=match_title
    df['Series']=series
    df['Place']=place
    df['Date']=date
    df['Time']=mat_time
    
    driver.close()
    
    return df


# Calling Function
cricket_fixtures('https://www.bcci.tv/')
```

## 3. Scrape the details of selenium exception from guru99.com.Url = https://www.guru99.com/. You need to find following details:Name, Description. Note: - From guru99 home page you have to reach to selenium exception handling page through code.

```python
# Function Definition
def Selenium_Exception(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     clicking selenium
    sele = []
    for i in driver.find_elements_by_xpath("//li[@class='fa fa-chevron-circle-right']/a"):
        sele.append(i)   
    for i in sele:
        if i.text == 'Selenium':
            i.click()
            break
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'table')))
    
#     Clicking Selenium Exception Handling Tutorial
    tutorials = []
    for i in driver.find_elements_by_xpath("//table[@class='table']/tbody/tr/td/a"):
        tutorials.append(i)
    for i in tutorials:
        if i.get_attribute('title')=='Selenium Exception Handling (Common Exceptions List)':
            url = i.get_attribute('href')
            driver.get(url)
            break
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'page-header')))
    
#     Let's create empty lists to store necessary information
    exception_name = []
    description = []
    
#     extracting exception_name
    for i in driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[1]"):
        exception_name.append(i.text)
    
#     extracting description
    for i in driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[2]"):
        description.append(i.text)
    
    df = pd.DataFrame({})
    df['Exception_Name']=exception_name[1:]
    df['Description']=description[1:]
    
    driver.close()
    
    return df


# Calling Function
Selenium_Exception('https://www.guru99.com/')
```

## 4. Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/. You have to find following details: Rank, State, GSDP(19-20), GSDP(18-19), Share, GDP(Billion). Note: - From statisticstimes home page you have to reach to economy page through code.

```python
# Function Definition
def GDP(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'dropbtn')))
    
#     clicking economy
    button = []
    for i in driver.find_elements_by_xpath("//button[@class='dropbtn']"):
        button.append(i)
    for i in button:
        if i.text=='Economy':
            i.click()
            break
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'dropdown-content')))
    
#     clicking India
    dropdown_content = []
    for i in driver.find_elements_by_xpath("//div[@class='dropdown-content']/a"):
        dropdown_content.append(i)
    for i in dropdown_content:
        if i.text=='India':
            url = i.get_attribute('href')
            driver.get(url)
            break
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'ec')))
    
#     clicking GDP of Indian states
    content = []
    for i in driver.find_elements_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li/a"):
        content.append(i)
    for i in content:
        if i.text=='» GDP of Indian states':
            url = i.get_attribute('href')
            driver.get(url)
            break
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'fwidth')))
    
#     extracting information
    info=[]
    for i in driver.find_elements_by_xpath("//table[@class='display dataTable']/tbody/tr/td"):
        info.append(i.text)
    
    df = pd.DataFrame({})
    df['Rank']=info[0:264:8]
    df['State']=info[1:264:8]
    df['GSDP(19-20)']=info[2:264:8]
    df['GSDP(18-19)']=info[3:264:8]
    df['Share(18-19)']=info[4:264:8]
    df['GDP($ Billion)']=info[5:264:8]
    
    driver.close()
    
    return df


# Calling Function
GDP('http://statisticstimes.com/')
```

## 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/. You have to find the following details: Repository title, Repository description, Contributors count, Language used. Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
# Function Definition
def github(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     clicking Explore
    button = []
    for i in driver.find_elements_by_xpath("//summary[@class='HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block']"):
        button.append(i)
    for i in button:
        if i.text=='Explore':
            i.click()
            break
    
#     clicking Trending
    trending=driver.find_element_by_xpath("//a[@class='py-2 lh-condensed-ultra d-block Link--secondary no-underline f5 Bump-link--hover' and @data-ga-click='(Logged out) Header, go to Trending']").get_attribute('href')
    driver.get(trending)
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//h1[@class='h3 lh-condensed']/a")))
    
#     extracting all repositories url
    repository_url=[]
    for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
        repository_url.append(i.get_attribute('href'))
    
#     Let's create empty lists to store necessary information
    repository_title=[]
    repository_description=[]
    contributors_count=[]
    language_used=[]

    for url in repository_url:
        driver.get(url)
        delay=10
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//strong[@class='mr-2 flex-self-stretch']/a")))
        
#         extracting repository_title
        rt=driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")
        repository_title.append(rt.text)
        
#         extracting repository_description
        try:
            rd=driver.find_element_by_xpath("//p[@class='f4 mt-3']")
            repository_description.append(rd.text)
        except NoSuchElementException:
            repository_description.append('-')
        
#         extracting contributors_count
        count=[]
        for i in driver.find_elements_by_xpath("//h2[@class='h4 mb-3']/a"):
            count.append(i.text)
        c=pd.Series(count)
        contributors_count.append(list(c[c.str.contains('Contributors')]))
        
#         extracting language_used
        language=[]
        for i in driver.find_elements_by_xpath("//span[@class='color-text-primary text-bold mr-1']"):
            language.append(i.text)
        language_used.append(", ".join(language))
    
    
    df=pd.DataFrame({})
    df['Repository_Title']=repository_title
    df['Repository_Description']=repository_description
    df['Contributors_Count']=contributors_count
    df['Language_Used']=language_used
    
    driver.close()
    
    return df


# Calling Function
github('https://github.com/')
```

## 6. Scrape the details of top 100 songs on billiboard.com. Url = https://www.billboard.com/. You have to find the following details: Song name, Artist name, Last week rank, Peak rank, Weeks on board. Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
# Function Definition
def songs(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     clicking charts
    charts = []
    for i in driver.find_elements_by_xpath("//a[@class='header__main-link header__main-link--charts']"):
        charts.append(i.get_attribute('href'))
    driver.get(charts[0])
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'charts-landing__info--view-chart-holder')))
    
#     clicking view_chart
    view_chart=driver.find_element_by_xpath("//div[@class='charts-landing__info--view-chart-holder']/div[1]")
    view_chart.click()
    
    delay=10
    WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CLASS_NAME,'chart-element__information')))
    
#     Let's create empty lists to store necessary information
    song_artist_name=[]
    last_week_rank=[]
    peak_rank=[]
    weeks_on_board=[]
    
#     extracting song_name and artist_name
    for i in driver.find_elements_by_xpath("//span[@class='chart-element__information']/span"):
        song_artist_name.append(i.text)
    
#     extracting last_week_rank
    for i in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']"):
        last_week_rank.append(i.text)
    
#     extracting peak_rank
    for i in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']"):
        peak_rank.append(i.text)
    
#     extracting weeks_on_board
    for i in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']"):
        weeks_on_board.append(i.text)
    
    df=pd.DataFrame({})
    df['Song_Name']=song_artist_name[::3]
    df['Artist_Name']=song_artist_name[1::3]
    df['Last_Week_Rank']=last_week_rank
    df['Peak_Rank']=peak_rank
    df['Weeks_On_Board']=weeks_on_board
    
    driver.close()
    
    return df


# Calling Function
songs('https://www.billboard.com/')    
```

## 7. Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/. You have to find the following details:Name, Designation, Company, Skills they hire for, Location. Note: - From naukri.com homepage click on the recruiters option and the on the search panel type Data science and click on search. All this should be done through code.

```python
# Function Definition
def jobs(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     clicking recruiters
    recruiters=driver.find_element_by_xpath("//a[@title='Search Recruiters']").get_attribute('href')
    driver.get(recruiters)
    time.sleep(6)
    
#     searching Data Science
    search_bar=driver.find_element_by_xpath("//div[@id='skill']/div/div[2]/input")
    search_bar.send_keys('Data Science')
    search_butn=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
    search_butn.click()
    time.sleep(6)
    
#     extracting all jobs url
    href=[]
    for i in driver.find_elements_by_xpath("//a[@class='ellipsis' and @target='_blank']"):
        href.append(i.get_attribute('href'))
    
#     Let's create empty lists to store necessary information
    name=[]
    designation=[]
    company_location=[]
    skills_required=[]

    for url in href:
        driver.get(url)
        delay=10
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//h1[@class='fl ellipsis wLimit hd']")))
        
#         extracting name
        try:
            n=driver.find_element_by_xpath("//h1[@class='fl ellipsis wLimit hd']")
            name.append(n.text)
        except NoSuchElementException:
            name.append("-")
        
#         extracting designation
        try:
            d=driver.find_element_by_xpath("//div[@class='ellipsis']")
            designation.append(d.text)
        except NoSuchElementException:
            designation.append("-")
        
#         extracting company and location
        try:
            for i in driver.find_elements_by_xpath("//div[@class='oh']/a"):
                company_location.append(i.text)
        except NoSuchElementException:
            company_location.append("-")

#         extracting skills_required
        try:
            s=driver.find_element_by_xpath("//div[@class='fl lPortn']/p[1]")
            skills_required.append(s.text)
        except NoSuchElementException:
            skills_required.append("-")
        
        df=pd.DataFrame({})
    df['Name']=name
    df['Designation']=designation
    df['Company']=company_location[::2]
    df['Skills_Required']=skills_required
    df['Location']=company_location[1::2]
    
    driver.close()
    
    return df


# Calling Function
jobs('https://www.naukri.com/')
```

## 9. Scrape the details of most watched tv series of all time from imdb.com.Url = https://www.imdb.com/list/ls095964455/. You have to find the following details:Name, Year span, Genre, Run time, Ratings, Votes.

```python
def tv_series(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     Let's create empty lists to store necessary information
    name =[]
    year_span =[]
    genre=[]
    run_time=[]
    ratings=[]
    votes=[]
    
#     extracting name
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
        name.append(i.text)
    
#     extracting year_span
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]"):
        year_span.append(i.text)
    
#     extracting genre
    for i in driver.find_elements_by_xpath("//span[@class='genre']"):
        genre.append(i.text)
    
#     extracting run_time
    for i in driver.find_elements_by_xpath("//span[@class='runtime']"):
        run_time.append(i.text)
    
#     extracting ratings
    for i in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]"):
        ratings.append(i.text)
    
#     extracting votes
    for i in driver.find_elements_by_xpath("//span[@name='nv']"):
        votes.append(i.text)
    
    df=pd.DataFrame({})
    df['Name']=name
    df['Year_Span']=year_span
    df['Genre']=genre
    df['Run_Time']=run_time
    df['Ratings']=ratings
    df['Votes']=votes
    
    driver.close()
    
    return df


# Calling Function
tv_series('https://www.imdb.com/list/ls095964455/')
```

## 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/. You have to find the following details: Dataset name, Data type, Task, Attribute type, No of instances,  No of attributes, Year. Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Function Definition
def uci_repositories(url):
    driver = webdriver.Chrome("C:\\chromedriver.exe")
    driver.get(url)
    time.sleep(8)
    
#     clicking View_ALL_Data_Sets
    View_ALL_Data_Sets=driver.find_element_by_xpath("//span[@class='whitetext']/a/font")
    View_ALL_Data_Sets.click()
    time.sleep(8)
    
#     Let's create empty lists to store necessary information
    name=[]
    data_type=[]
    task=[]
    attribute_type=[]
    no_of_instances=[]
    no_of_attributes=[]
    year=[]
    
#     extracting all Data_Sets url
    href=[]
    for i in driver.find_elements_by_xpath("//b/a"):
        href.append(i.get_attribute('href'))
    
    for url in href:
        driver.get(url)
        try:
            delay=10
            WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"//span[@class='heading']/b")))
        except TimeoutException:
            continue
        
#         extracting name
        try:
            for i in driver.find_elements_by_xpath("//span[@class='heading']/b"):
                name.append(i.text)
        except NoSuchElementException:
            name.append('-')
        
#         extracting data_type
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[2]/p"):
                content.append(i.text)
            data_type.append(content[0])
        except NoSuchElementException:
            data_type.append('-')
        
#         extracting task
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[2]/p"):
                content.append(i.text)
            task.append(content[2])
        except NoSuchElementException:
            task.append('-')
        
#         extracting attribute_type
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[2]/p"):
                content.append(i.text)
            attribute_type.append(content[1])
        except NoSuchElementException:
            attribute_type.append('-')

#         extracting no_of_instances
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[4]/p"):
                content.append(i.text)
            no_of_instances.append(content[0])
        except NoSuchElementException:
            no_of_instances.append('-')
        
#         extracting no_of_attributes
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[4]/p"):
                content.append(i.text)
            no_of_attributes.append(content[1])
        except NoSuchElementException:
            no_of_attributes.append('-')
        
#         extracting year
        try:
            content=[]
            for i in driver.find_elements_by_xpath("//tr/td[6]/p"):
                content.append(i.text)
            year.append(content[1])
        except NoSuchElementException:
            year.append('-')
    
    df=pd.DataFrame({})
    df['Name']=name
    df['Data_Type']=data_type
    df['Task']=task
    df['Attribute_Type']=attribute_type
    df['No_of_Instances']=no_of_instances
    df['No_of_Attributes']=no_of_attributes
    df['Year']=year
    
    driver.close()
    
    return df


# Calling Function
uci_repositories('https://archive.ics.uci.edu/')
```
